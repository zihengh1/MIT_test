```python
import numpy as np
import pandas as pd
import csv
import datetime
import time
import random
import os
import matplotlib.pyplot as plt
```

```python
def deal_with(data_in):
    collect = {}
    for x in data_in:
        h = int(x[1]) + 8
        if h >= 24:
            h = h - 24
        time = str(h) + ":" + str(x[2]) + ":" + str(x[3])
        collect[time] = int(x[0])
    return collect
```

```python
def delete(data_in):
    minus = []
    for x in data_in[0:len(data_in):10]:
        minus.append(x)
    return minus
```

```python
def open_file(path):
    d_temp = []
    ans = []
    with open(path) as csvDataFile:
        csvReader = csv.reader(csvDataFile)
        for row in csvReader:
            d_temp.append(row)
        for x in d_temp[0:len(d_temp):10]:
            ans.append(x)
        data = deal_with(ans)
        return data
```

```python
path1 = '/Users/heng/Desktop/3:27 18~3:28 18/47_data_out'
first = open_file(path1)
path2 = '/Users/heng/Desktop/3:27 18~3:28 18/99_data_out'
second = open_file(path2)
# print(len(list(first)))
# print(len(list(second)))
M = min(len(list(first)), len(list(second)))
```

```python
df = pd.DataFrame({"47_d": np.array(list(first.values())),
                   "99_d": np.array(list(second.values()))}, index = np.array(list(second.keys())))
```

```python
df = df.astype(float)
df.plot()
plt.title('3/26 23 ~ 3/27 7')
plt.grid(True)
plt.show()
```

```python
total_std = []
average = []
for x in range(0, 267):
    row = np.array(df.iloc[x])
    std = row.std()
    avg = row.mean()
    total_std.append(std)
    average.append(avg)
total_std = np.array(total_std)
average = np.array(average)
CV = total_std / average
```

```python
df = pd.DataFrame({"CV": CV}, index = average)
```

```python
plt.scatter(average, CV)
plt.grid(True)
plt.show()
```
